```python
import numpy as np

# Generate two 20x20 matrices with random numbers between 0 and 1000
A = np.random.randint(0, 1001, size=(20, 20))
B = np.random.randint(0, 1001, size=(20, 20))

# Compute the matrix multiplication C = AB
C = np.dot(A, B)

# Print the resulting matrix C
print("Matrix C:")
print(C)
```

```python
A = [[1, 0, 2], [-1, 3, 1], [0, 2, 1]]
B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]
# Initialize the result matrix with zeros
result = [[0, 0, 0],
          [0, 0, 0],
          [0, 0, 0]]

# Perform matrix multiplication
for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(A)):
            result[i][j] += A[i][k] * B[k][j]
print(result)
```

```python
A = [[1, 0, 2], [-1, 3, 1], [0, 2, 1]]
B = [1, 2, 3]  

# Initialize the result vector with zeros
result = [0, 0, 0]

# Perform matrix-vector multiplication
for i in range(len(A)):
    for j in range(len(B)):
        result[i] += A[i][j] * B[j]

print(result)
```
